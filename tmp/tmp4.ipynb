```python
import keras 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras import layers
import cv2
import numpy as np
import tensorflow as tf


tf.random.set_seed(0)

img = cv2.imread('image.png',cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (10, 10))

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()




model=keras.Sequential()
model.add(layer=layers.Conv2D(input_shape=(10,10,1),filters=64,kernel_size=(3,3)))
model.add(layers.MaxPool2D(pool_size=(3,3),strides=(2,2)))


feature_map=model.predict(np.array([img]))
# print(feature_map.shape)


feature_img=feature_map[0,:,:,0]
plt.imshow(feature_img,cmap='gray') 


# for i in range(10):
#     feature_img=feature_map[0,:,:,i]
#     ax=plt.subplot(8,8,i+1)
#     plt.imshow(feature_img)


```
